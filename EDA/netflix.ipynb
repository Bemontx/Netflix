## Verificacion Cantidad Null

```python
print(df.shape)
df.isnull().sum()
```

```python
def nulos(df):
    porcentajes_nulos = (df.isnull().mean() * 100).round(2)
    resumen_nulos = pd.DataFrame({'Columna': porcentajes_nulos.index, 'Porcentaje Nulos': porcentajes_nulos.values})
    resumen_nulos = resumen_nulos.sort_values(by='Porcentaje Nulos', ascending=False)

    print(resumen_nulos)

nulos(df)
```

## Promedio generos y calcula el ingreso mensual promedio por paises

```python
grouped_data = df.groupby(['Country', 'Gender']).agg({'Monthly Revenue': 'mean'})
grouped_data
```

```python
# Datos
countries = ['Australia', 'Brazil', 'Canada', 'France', 'Germany', 'Italy', 'Mexico', 'Spain', 'United Kingdom', 'United States']
genders = ['Female', 'Male']
monthly_revenue = [
    [12.370787, 12.446809],
    [12.168421, 12.829545],
    [12.566879, 12.356250],
    [12.593407, 12.619565],
    [12.212766, 12.494382],
    [12.615385, 12.706522],
    [12.258427, 12.191489],
    [12.562232, 12.545872],
    [12.752688, 12.577778],
    [12.711111, 12.407080]
]

# Crear el gráfico
fig = go.Figure()

for i, country in enumerate(countries):
    for j, gender in enumerate(genders):
        fig.add_trace(go.Bar(
            x=[country],
            y=[monthly_revenue[i][j]],
            name=gender,
            marker_color='blue' if gender == 'Female' else 'orange'
        ))

# Ajustes del diseño
fig.update_layout(
    title='Ingresos Mensuales Promedio por País y Género',
    xaxis=dict(title='País'),
    yaxis=dict(title='Ingreso Mensual Promedio'),
    barmode='group',
    legend=dict(title='Género')
)

# Mostrar el gráfico
fig.show()
```

### Convertir columna fecha a su tipo de dato para graficacion de inscripciones por fechas

```python
df['Join Date'] = pd.to_datetime(df['Join Date'], format='%d-%m-%y')
df['Last Payment Date'] = pd.to_datetime(df['Last Payment Date'], format='%d-%m-%y')
```

### Inscripciones por mes

```python
def asignar_semestre(fecha):
    if fecha.month <= 6:
        return '1er Semestre'
    else:
        return '2do Semestre'

df['Semestre de Join'] = df['Join Date'].apply(asignar_semestre)

inscripciones_por_semestre = df.groupby('Semestre de Join').size().reset_index(name='Inscripciones')

fig = px.bar(inscripciones_por_semestre, x='Semestre de Join', y='Inscripciones', title='Inscripciones de usuarios por semestre de inscripción')
fig.show()
```

### Asignacion de coordenadas por region de cada Pais

```python
coordinates = {
    'Country': ['Australia', 'Brazil', 'Canada', 'France', 'Germany', 'Italy', 'Mexico', 'Spain', 'United Kingdom', 'United States'],
    'Latitude': [-25.2744, -14.2350, 56.1304, 46.6034, 51.1657, 41.8719, 23.6345, 40.4637, 55.3781, 37.7749],
    'Longitude': [133.7751, -51.9253, -106.3468, 1.8883, 10.4515, 12.5674, -102.5528, -3.7492, -3.4360, -122.4194]
}
coordinates_df = pd.DataFrame(coordinates)

df_with_coordinates = pd.merge(df, coordinates_df, on='Country')

fig_map = px.density_mapbox(df_with_coordinates, lat='Latitude', lon='Longitude', z='Monthly Revenue', radius=10,
                            center=dict(lat=0, lon=180), zoom=0, mapbox_style="carto-positron",
                            title='Distribución de Monthly Revenue por país')
fig_map.show()
```
